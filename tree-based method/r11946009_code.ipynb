# Preprocessing


## cusinfo (len: 24)
### Categorical
- occupation_code
  - one-hot encoding (0 ~ 21) no drop_first -> **because nan will be all zeros**

```python
for i in range(21):
  pub_tr_x_custinfo[f"occupation_{i}"] = pd.get_dummies(pub_tr_x_custinfo.occupation_code)[float(i)]

pub_tr_x_custinfo = pub_tr_x_custinfo.drop(columns=['occupation_code'])
```

```python
pub_tr_x_custinfo.head()
```

## ccba (byymm)
- all numeric

```python
pub_tr_x_ccba.head()
```


## cdtx (date)

### Categorical
- country → **new_country**
  - 1: 130 (台灣)
  - 0: not 130
- cur_type → **new_cur_type**
  - 1: 47 (台幣)
  - 0: not 47 


```python
dic = {True: 1, False: 0}
pub_tr_x_cdtx['new_country'] = [dic[i] for i in pub_tr_x_cdtx.country == 130]
pub_tr_x_cdtx['new_cur_type'] = [dic[i] for i in pub_tr_x_cdtx.cur_type == 47]

# drop columns
pub_tr_x_cdtx = pub_tr_x_cdtx.drop(columns=['country', 'cur_type'])
```

```python
pub_tr_x_cdtx.head()
```

## dp (tx_date)

### Categorical
- debit_credit → **debit_credit**
  - 1: CR
  - 0: DB
- tx_type & info_asset_code → **new_type**
  - 1: type = 1 & info_asset_code = 12
  - 0: OW
- crossbank
  - 1: 跨行交易
  - 0: 非跨行
- ATM
  - 1: 實體 ATM 交易
  - 0: 非實體 ATM

### Delete
- tx_time
- fiscTxId
- txbranch

```python
# debit_credit
dic = {'CR': 1, 'DB': 0}
pub_tr_x_dp['debit_credit'] = [dic[i] for i in pub_tr_x_dp.debit_credit]

# new_type
ind = pub_tr_x_dp[pub_tr_x_dp.tx_type == 1][pub_tr_x_dp.info_asset_code == 12].index
new_type = []
for i in range(len(pub_tr_x_dp)):
  if i in ind:
    new_type.append(1)
  else:
    new_type.append(0)

pub_tr_x_dp['new_type'] = new_type

# drop columns
pub_tr_x_dp = pub_tr_x_dp.drop(columns=['tx_time', 'txbranch', 'fiscTxId', 'tx_type', 'info_asset_code'])
```

```python
pub_tr_x_dp.head()
```

## remit (trans_date)
### Categorical 
- trans_no
  - one-hot encoding (1, 2, 3, 4) drop_first = True

```python
pub_tr_x_remit['trans_no1'] = pd.get_dummies(pub_tr_x_remit.trans_no, drop_first = True)[1]
pub_tr_x_remit['trans_no2'] = pd.get_dummies(pub_tr_x_remit.trans_no, drop_first = True)[2]
pub_tr_x_remit['trans_no3'] = pd.get_dummies(pub_tr_x_remit.trans_no, drop_first = True)[3]
pub_tr_x_remit['trans_no4'] = pd.get_dummies(pub_tr_x_remit.trans_no, drop_first = True)[4]

pub_tr_x_remit = pub_tr_x_remit.drop(columns=['trans_no'])
```

```python
pub_tr_x_remit.head()
```

# Table 1-1、 Create Index Table

```python
# dic = {
#     cus_id: {
#         month: {
#             month_num: {ccba: []}
#         }
#         date: {
#             date_num: {
#                 cdtx: []
#                 dp: []
#                 remit: []                                
#             }
#         }
#     }
# }
```

```python
cus_list = pub_tr_x_custinfo.cust_id.unique()
ccba = pub_tr_x_ccba.to_dict()
cdtx = pub_tr_x_cdtx.to_dict()
dp = pub_tr_x_dp.to_dict()
remit = pub_tr_x_remit.to_dict()
```

```python
# create empty dictionary
dic = {}
for i in cus_list:
  dic[i] = {'month': {}, 'date': {}}
  for j in pub_tr_x_ccba.byymm.unique():
    dic[i]['month'][j] = {}
  for j in range(395):
    dic[i]['date'][j] ={}
```

## month data | ccba (len: 8)
- 'lupay', 'cycam', 'usgam', 'clamt', 'csamt', 'inamt', 'cucsm', 'cucah'

```python
progress = tqdm(total = len(ccba['cust_id']))
for ind in range(len(ccba['cust_id'])):
  temp = []
  for col in ccba.keys():
    if col not in ['cust_id', 'byymm']:
      temp.append(ccba[col][ind])
  dic[ccba['cust_id'][ind]]['month'][ccba['byymm'][ind]]['ccba'] = temp
  progress.update(1)
```

## date data | cdtx (len: 3)
- 'amt', 'new_country', 'new_cur_type'

```python
progress = tqdm(total = len(cdtx['cust_id']))
for ind in range(len(cdtx['cust_id'])):
  temp = []
  for col in cdtx.keys():
    if col not in ['cust_id', 'date']:
      temp.append(cdtx[col][ind])
  dic[cdtx['cust_id'][ind]]['date'][cdtx['date'][ind]]['cdtx'] = temp
  progress.update(1)
```

## date data | dp (len: 6)
- 'debit_credit', 'tx_amt', 'exchg_rate', 'cross_bank', 'ATM', 'new_type'

```python
progress = tqdm(total = len(dp['cust_id']))
for ind in range(len(dp['cust_id'])):
  temp = []
  for col in dp.keys():
    if col not in ['cust_id', 'tx_date']:
      temp.append(dp[col][ind])
  dic[dp['cust_id'][ind]]['date'][dp['tx_date'][ind]]['dp'] = temp
  progress.update(1)
```

## date data | remit (len: 5)
- 'trade_amount_usd', 'trans_no1', 'trans_no2', 'trans_no3', 'trans_no4'

```python
progress = tqdm(total = len(remit['cust_id']))
for ind in range(len(remit['cust_id'])):
  temp = []
  for col in remit.keys():
    if col not in ['cust_id', 'trans_date']:
      temp.append(remit[col][ind])
  dic[remit['cust_id'][ind]]['date'][remit['trans_date'][ind]]['remit'] = temp
  progress.update(1)
```

# Table 1-2、 Pad zeros

```python
progress = tqdm(total = len(cus_list))
for cus in cus_list:
  for month in pub_tr_x_ccba.byymm.unique():
    if 'ccba' not in dic[cus]['month'][month]:
      dic[cus]['month'][month]['ccba'] = [0]*8
  for date in range(395):
    if 'cdtx' not in dic[cus]['date'][date]:
      dic[cus]['date'][date]['cdtx'] = [0]*3
    if 'dp' not in dic[cus]['date'][date]:
      dic[cus]['date'][date]['dp'] = [0]*6
    if 'remit' not in dic[cus]['date'][date]:
      dic[cus]['date'][date]['remit'] = [0]*5
  progress.update(1)
```

```python
# with open('/content/drive/MyDrive/嘿嘿嘿 ver.2/FinTech/pkl/table1-2_origin_padzero.pkl', 'wb') as f:
#   pickle.dump(dic, f)
```

# Table 1-3、 Combine Big Index Table

```python
# dic = {
#     cus_id: {
#         date(named: "d_{date_num}") / month(named: "m_{month_num}"): {
#             if date: [cdtx(3), dp(6), remit(5)] dim = (1, 14)
#             if month: [ccba(8)]          dim = (1, 8)
#         }
#     }
# }
```

```python
# create empty dictionary
combined_dic = {}
progress = tqdm(total = len(cus_list))
for cus in cus_list:
  combined_dic[cus] = {}
  for month in pub_tr_x_ccba.byymm.unique():
    combined_dic[cus][f"m_{month}"] = dic[cus]['month'][month]['ccba']
  for date in range(395):
    temp = dic[cus]['date'][date]['cdtx'].copy()
    temp.extend(dic[cus]['date'][date]['dp'])
    temp.extend(dic[cus]['date'][date]['remit'])
    combined_dic[cus][f"d_{date}"] = temp
  progress.update(1)
```

## date

```python
def get_train_date_data(train_data, train_dic, combined_dic, n_date, which_way):
  """
  hyperparameter
    - which_way (also contain alert_date itself)
      - 1: only past data
      - 2: past data + future data
      - 3: only future data
  """
  progress = tqdm(total = len(train_data['alert_key']))
  for ind in range(len(train_data['alert_key'])):
    alert_date = train_data['date'][ind]
    temp = []

    # for alert_date itself & (n_date-1) future data
    if which_way == 3:
      for date_no in range(n_date):
        temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{alert_date+date_no}"])

    # for alert_date itself & (n_date-1) past data
    elif which_way == 1:
      if alert_date < n_date - 1:
        diff = n_date - alert_date
        for date_no in range(alert_date):
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{date_no}"])
        for times in range(diff):
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{alert_date}"])
      else:
        for date_no in range(n_date):
          date_no = n_date - date_no - 1
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{alert_date-date_no}"])

    # for alert itself & (n_date-1)/2 future data & (n_date-1)/2 past data
    elif which_way == 2:
      future_num = round((n_date-1)/2)
      past_num = n_date - 1 - future_num  
      if alert_date < past_num - 1 :
        diff = past_num - alert_date
        for date_no in range(alert_date):
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{date_no}"])
        for times in range(diff):
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{alert_date}"])
      else:
        for date_no in range(past_num):
          date_no = past_num - date_no
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{alert_date-date_no+1}"])
      for date_no in range(future_num+1):
          temp.extend(combined_dic[train_data['cust_id'][ind]][f"d_{alert_date+date_no}"])

    train_dic[train_data['alert_key'][ind]]['date'] = temp
    progress.update(1)

  return train_dic
```

## date

```python
def get_test_date_data(test_data, test_dic, combined_dic, n_date, which_way):
  """
  hyperparameter
    - which_way (also contain alert_date itself)
      - 1: only past data
      - 2: past data + future data
      - 3: only future data
  """
  max_date = 394
  progress = tqdm(total = len(test_data['alert_key']))
  for ind in range(len(test_data['alert_key'])):
    alert_date = test_data['date'][ind]
    temp = []

    # for alert_date itself & (n_date-1) future data
    if which_way == 3:
      if max_date - alert_date < n_date - 1:
        diff = n_date - (max_date - alert_date)
      for date_no in range(max_date-alert_date):
        temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{alert_date+date_no}"])
      for times in range(diff):
        temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{max_date}"])
    
    # for alert_date itself & (n_date-1) past data
    elif which_way == 1:
      for date_no in range(n_date):
        date_no = n_date - date_no - 1
        temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{alert_date-date_no}"])

    # # for alert itself & (n_date-1)/2 future data & (n_date-1)/2 past data
    elif which_way == 2:
      future_num = round((n_date-1)/2)
      past_num = n_date - 1 - future_num
      
      # past_data
      for date_no in range(past_num):
        date_no = past_num - date_no
        temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{alert_date-date_no}"])
      # future_data
      if alert_date + future_num > max_date:
        diff = alert_date + future_num - max_date + 1
        for date_no in range(max_date-alert_date):
          temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{alert_date+date_no}"])
        for times in range(diff):
          temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{max_date}"])
      else:
        for date_no in range(future_num + 1):
          temp.extend(combined_dic[test_data['cust_id'][ind]][f"d_{alert_date+date_no}"])
      
    test_dic[test_data['alert_key'][ind]]['date'] = temp
    progress.update(1)

  return test_dic
```
